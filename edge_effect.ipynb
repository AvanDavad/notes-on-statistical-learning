## curse of dimensionality and the edge effect problem

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
def plot_ndim(n=3, num_points=100):
    x = np.random.randn(num_points,n)
    x2 = project_to_plane(x)
    scatter(x2)
    plt.title(f"{n} dimension")
```

```python
def project_to_plane(x):
    x_abs = np.sqrt((x**2).sum(axis=1, keepdims=True))
    x2_abs = np.sqrt((x[:,:2]**2).sum(axis=1, keepdims=True))
    factor = x_abs / x2_abs
    x2 = x[:,:2]*factor
    return x2
```

```python
def scatter(x):
    plt.scatter(x[:,0], x[:,1], s=1)
    side_len = np.max(np.abs(x))
    plt.xlim(-side_len, side_len)
    plt.ylim(-side_len, side_len)
    plt.gca().set_aspect('equal', adjustable='box')
```

### distances from origin, in n dimension

```python
plot_ndim(2, 1000)
```

## setting a random sample point as origin

```python
def plot_ndim_rel(ndim=2, num_points=1000):
    x = np.random.randn(num_points, ndim)
    x -= x[0:1]
    x2 = project_to_plane(x[1:])
    scatter(x2)
    plt.scatter([0], [0], c='r')
    plt.title(f"{ndim} dimension")
```

```python
plot_ndim_rel(2)
```

```python
plot_ndim_rel(5)
```

```python
plot_ndim_rel(10)
```

```python
plot_ndim_rel(15)
```
